```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Anshu Kumar\OneDrive\Documents\project\sugar\Diabetes dataset.csv")
# print(df.head())
df
```

```python
# df.info()
print(df.isnull().sum())
```

```python
for col in ['Age', 'BMI', 'Glucose', 'BloodPressure']:
    df[col] = df[col].fillna(df[col].median())  # Assign back so it updates

df['Diabetic'] = df['Diabetic'].replace(r'^\s*$', np.nan, regex=True)


# Drop rows where 'Diabetic' is missing
df = df.dropna(subset=['Diabetic']).copy()
# # Check if nulls remain
print(df.isnull().sum())
# print(df.head())   
```
